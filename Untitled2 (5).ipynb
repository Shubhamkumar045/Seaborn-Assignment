```python
Que 1: Name any five plots that we can plot using the Seaborn library.
Answer--Also, state the uses of each plot.
Answer--Seaborn is a powerful Python visualization library based on Matplotlib. 
It provides a high-level interface for drawing attractive and informative statistical graphics.
Seaborn offers various types of plots to explore and visualize data effectively. 
Here are five common plots that can be created using Seaborn along with their uses:

E)Scatter Plot:
Use: Scatter plots are used to visualize the relationship between two continuous variables. 
They help in identifying patterns, clusters, or trends in the data.
A)Histogram:
Use: Histograms are used to visualize the distribution of a single continuous variable. 
They display the frequency or probability density of data within predefined bins or intervals.
B)Bar Plot:
Use: Bar plots are used to compare the values of categorical variables. 
They represent the frequency, count, or proportion of data points in different categories.
C)Box Plot:
Use: Box plots (box-and-whisker plots) provide a visual summary of the 
distribution of a continuous variable. They display the median, quartiles, 
and range of the data, as well as identify outliers.
D)Heatmap:
Use: Heatmaps are used to visualize the relationship between two categorical
variables or between a categorical and a continuous variable. They display data in a grid where 
each cell is color-coded based on the value it represents, making it easier to identify patterns
or correlations in large datasets.

Que 2: Load the "fmri" dataset using the load_dataset function of seaborn. Plot a line plot using x =
"timepoint" and y = "signal" for different events and regions.
Note: timepoint, signal, event, and region are columns in the fmri dataset.
Answer--To accomplish this task, we need to follow these steps:

Load the "fmri" dataset using Seaborn's load_dataset function.
Extract the required columns: "timepoint", "signal", "event", and "region".
Plot a line plot using "timepoint" as the x-axis and "signal" as the y-axis,
while differentiating the lines based on the "event" and "region" columns.

import seaborn as sns
import matplotlib.pyplot as plt

# Load the fmri dataset
fmri_data = sns.load_dataset("fmri")

# Plot line plot
plt.figure(figsize=(10, 6))

# Iterate over unique combinations of event and region
for (event, region), data in fmri_data.groupby(['event', 'region']):
    plt.plot(data['timepoint'], data['signal'], label=f'{event}-{region}')

# Adding title and labels
plt.title('FMRI Signal over Time for Different Events and Regions')
plt.xlabel('Timepoint')
plt.ylabel('Signal')

# Adding legend
plt.legend()

# Show the plot
plt.grid(True)
plt.tight_layout()
plt.show()


Que 3: Load the "titanic" dataset using the load_dataset function of seaborn. Plot two box plots using x =
'pclass', y = 'age' and y = 'fare'.
Answer--To accomplish this task, we need to follow these steps:

Load the "titanic" dataset using Seaborn's load_dataset function.
Extract the required columns: "pclass", "age", and "fare".
Plot two box plots: one for "age" and one for "fare", with "pclass" on the x-axis.

import seaborn as sns
import matplotlib.pyplot as plt

# Load the titanic dataset
titanic_data = sns.load_dataset("titanic")

# Plot box plots
plt.figure(figsize=(12, 6))

# Box plot for age
plt.subplot(1, 2, 1)
sns.boxplot(x='pclass', y='age', data=titanic_data)
plt.title('Box Plot of Age by Pclass')

# Box plot for fare
plt.subplot(1, 2, 2)
sns.boxplot(x='pclass', y='fare', data=titanic_data)
plt.title('Box Plot of Fare by Pclass')

# Adjust layout
plt.tight_layout()

# Show the plot
plt.show()


Que 4: Use the "diamonds" dataset from seaborn to plot a histogram for the 'price' column. Use the hue
parameter for the 'cut' column of the diamonds dataset.
Answer--To plot a histogram for the 'price' column of the "diamonds" dataset from 
Seaborn while using the 'cut' column as the hue parameter, you can follow these steps:

Load the "diamonds" dataset using Seaborn's load_dataset function.
Use Seaborn's histplot function to create the histogram, specifying the 
'price' column for the x-axis and the 'cut' column for the hue parameter.
```
